```python
def FiniteDifference(Xmin,Xmax,Ymin,Ymax,dx,dy,Vleft,Vright,Vbottom,Vtop,d,w,p):
    """Returns the table of potential values and number of iterations 
    ran for given conditions: Xmin, Xmax, Ymin, Ymax, dx, dy, Vleft, 
    Vright, Vbottom, Vtop, Iteration Criteria, Overrelaxation Parameter, 
    and Charge Density."""
    x=arange(Xmin,Xmax+dx,dx)
    y=arange(Ymin,Ymax+dy,dy)
    Count=0
    Grid=zeros((size(x),size(y)),float)
    for i in range(size(x)):
        for j in range(size(y)):
            Grid[i,j]=(Vtop+Vbottom+Vleft+Vright)/float(4)
    for i in range(size(x)):
        Grid[i,size(y)-1]=Vtop
        Grid[i,0]=Vbottom
    for i in range(size(y)):
        Grid[size(x)-1,i]=Vright
        Grid[0,i]=Vleft
    dn=1 #new iteration condition
    do=1 #old iteration condition
    IC=1 #(Vnew-Vold)/Vold
    ICo=0 #old IC
    e0=double(8.854187817*10**-12)
    if size(p)==1:
        while abs(IC)>=d or abs(ICo)<abs(IC):
        #for k in range(1239):
            Count+=1
            ICo=(dn-do)/do
            do=Grid[float(size(x))/2,float(size(y))/2]
            for i in range(1,size(x)-1):
                for j in range(1,size(y)-1):
                    Grid[i,j]=w*(Grid[i+1,j]+Grid[i-1,j]+Grid[i,j+1]+Grid[i,j-1])/4+(1-w)*Grid[i,j]
            dn=Grid[float(size(x))/2,float(size(y))/2]
            IC=(dn-do)/do
        return Grid, Count
    else:
        #while abs(IC)>=d or abs(ICo)<abs(IC):
        for k in range(1031):
            Count+=1
            ICo=(dn-do)/do
            do=Grid[float(size(x))/2,float(size(y))/2]
            for i in range(1,size(x)-1):
                for j in range(1,size(y)-1):
                    Grid[i,j]=(Grid[i+1,j]+Grid[i-1,j]+Grid[i,j+1]+Grid[i,j-1])/4+p[i,j]*dx*dy/(4*e0)
            dn=Grid[float(size(x))/2,float(size(y))/2]
            IC=(dn-do)/do
        return Grid, Count
        
```

```python
#test case 1
t=time.time()
x=arange(0,11.1,.1)
y=arange(0,6.1,.1)
Count=0
w=1
d=10**-6
dn=1
do=1
IC=1
ICo=0
Vtop=0
Vbottom=1
Vleft=0
Vright=0
Grid=zeros((size(x),size(y)),float)
for i in range(size(x)):
    for j in range(size(y)):
        Grid[i,j]=(Vtop+Vbottom+Vleft+Vright)/float(4)
for i in range(size(x)):
    Grid[i,size(y)-1]=Vtop
    Grid[i,0]=Vbottom
for i in range(size(y)):
    Grid[size(x)-1,i]=Vright
    Grid[0,i]=Vleft
while IC>=d or ICo<IC:
    Count+=1
    ICo=(dn-do)/do
    do=Grid[float(size(x))/2,float(size(y))/2]
    for i in range(1,size(x)-1):
        for j in range(1,size(y)-1):
            Grid[i,j]=w*(Grid[i+1,j]+Grid[i-1,j]+Grid[i,j+1]+Grid[i,j-1])/4+(1-w)*Grid[i,j]
    dn=Grid[float(size(x))/2,float(size(y))/2]
    IC=(dn-do)/do
#print str(IC)
#print Grid
print Count
X,Y=meshgrid(y,x)
fig = plt.figure(figsize=(8,6))

ax = fig.add_subplot(1, 1, 1, projection='3d')

p = ax.plot_surface(X,Y,Grid,cstride=2,rstride=2,cmap=cm.jet)
ellapsed=time.time()-t
print ellapsed
```

```python
def V(l,x,y):
    sol=0
    for n in range(1,l+1,1):
        sol+=(-2*(1-cos(n*pi))/(n*pi*tanh(n*pi*6/11))*sin(n*pi*x/11)*(sinh(n*pi*y/11)-tanh(n*pi*6/11)*cosh(n*pi*y/11)))
    return sol
```

```python
t,z=meshgrid(y,x)
Analytical=V(19,z,t)
Diff=Analytical-Ex1_C6[0]
```
