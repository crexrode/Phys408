```python
def V(l,x,y):
    sol=0
    for n in range(1,l+1,1):
        sol+=(-2*(1-cos(n*pi))/(n*pi*tanh(n*pi*6/11))*sin(n*pi*x/11)*(sinh(n*pi*y/11)-tanh(n*pi*6/11)*cosh(n*pi*y/11)))
    return sol
```
